# Introduction to RNN Forward Propagation

Recurrent Neural Networks (RNNs) are designed to handle **sequential data** such as text, speech, and time series. Unlike traditional feedforward networks, RNNs retain **memory of past inputs** through hidden states, making them effective for tasks where context matters.

In forward propagation:
- Inputs are processed **one step at a time**.
- Each step combines the **current input** with the **previous hidden state**.
- Shared weights across time steps ensure consistency.
- Outputs capture both **present information** and **historical context**.

This mechanism allows RNNs to model dependencies in sequences, such as predicting sentiment in a movie review or forecasting future values in a time series.


```python
from keras import Sequential
from keras.layers import Dense, SimpleRNN
```

```python
model = Sequential()

model.add(SimpleRNN(3,input_shape=(4,5)))
model.add(Dense(1,activation="sigmoid"))
model.summary()
```

```python
from keras.utils import plot_model
plot_model(model, show_shapes=True)
```

```python
print(model.get_weights()[0].shape)
print(model.get_weights())
```
