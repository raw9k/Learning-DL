```python
datagen = ImageDataGenerator(
    rotation_range=30,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2
)
```

```python
img1 = img_to_array(img)
```

```python
input_batch = img1.reshape(1,200,200,3)
```

```python
i =0
for output in datagen.flow(input_batch, batch_size=1,save_to_dir=r"C:\Users\rauna\codes\datasets\catsvsdogs\data_aug"):
    i = i +1
    if i ==10:
        break
```

```python
batch_size = 16

# prepare data augmentation configuration
train_datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)

test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    directory=r"C:\Users\rauna\codes\datasets\catsvsdogs\train",
    target_size=(150, 150),
    batch_size=batch_size,
    class_mode='binary')
```

```python
validation_generator = test_datagen.flow_from_directory(
    directory=r"C:\Users\rauna\codes\datasets\catsvsdogs\test",
    target_size=(150, 150),
    batch_size=batch_size,
    class_mode='binary')
```
