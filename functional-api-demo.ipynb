```python
from keras.models import Model
```

```python
from keras.layers import *
```

```python
x = Input(shape=(3,))
```

```python
hidden1 = Dense(128, activation="relu")(x)
hidden2 = Dense(64,activation='relu')(hidden1)

output1= Dense(1,activation="linear")(hidden2)
output2 = Dense(1, activation="sigmoid")(hidden2)
```

```python
model = Model(inputs=x, outputs = [output1,output2])
```

```python
model.summary()
```

```python
from keras.utils import plot_model
plot_model(model, show_shapes=True)
```

```python
inputA = Input(shape=(32,))
inputB = Input(shape=(128,))
```

```python
hidden_A1 = Dense(8,activation="relu")(inputA)
hidden_A2 = Dense(4,activation="relu")(hidden_A1)
```

```python
hidden_B1 = Dense(64,activation="relu")(inputB)
hidden_B2 = Dense(32,activation="relu")(hidden_B1)
hidden_B3 = Dense(4,activation="relu")(hidden_B2)
```

```python
combined = concatenate([hidden_A2,hidden_B3])
```

```python
x= Dense(2,activation="relu")(combined)
output = Dense(1,activation="sigmoid")(x)
```

```python
model2 = Model(inputs = [inputA,inputB], outputs = output)
```

```python
plot_model(model2, show_shapes=True,show_layer_names=True, dpi=50)
```
